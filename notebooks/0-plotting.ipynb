```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("../local-data/output/titanic_train_clean.csv")
df.shape
```

```python
cols = ['Survived', 'Pclass', 'Sex', 'FamilySize', 'hasCabin', 'isChild']
for col in cols:
    df[col] = df[col].astype('category')
df['Age2'] = df['Age2'].astype(int)
df.info()
```

```python
df_agg = df.groupby(['Survived', 'Sex'])['Survived'].count().unstack()
df_agg.plot(kind='bar', stacked=True) ;
```

```python
df_agg = df.groupby(['Survived', 'hasCabin'])['Survived'].count().unstack()
df_agg.plot(kind='bar', stacked=True) ;
```

```python
df_agg = df.groupby(['Survived', 'Pclass'])['Survived'].count().unstack()
df_agg.plot(kind='bar', stacked=True) ;
```

```python
df_agg = df.groupby(['Survived', 'FamilySize'])['Survived'].count().unstack()
df_agg.plot(kind='bar', stacked=True) ;
```
